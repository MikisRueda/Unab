```python
#Exponencial
import numpy as np
from scipy . optimize import curve_fit
import matplotlib . pyplot as plt

x = np . array ([2 ,4 ,6 ,8 ,10 ,12])
y = np . array ([2.2 ,3.0 ,4.5 ,6.0 ,8.5 ,12])

xs = np . linspace (0 ,15 ,15)

def f (x ,a , b ) :
  return a * np . exp ( - b * x )

param , _ = curve_fit (f ,x ,y , p0 =[0.1 , -0.0001])
print ( param )

fig = plt . figure ()
plt . scatter (x , y , label ='Datos ')
plt . plot ( xs , f ( xs ,* param ) , ':r', label ='Regresi√≥n exponencial ')
plt . xlabel ('X ')
plt . ylabel (' Y ')
plt . legend ()
plt . title ('Taller 17 ')
plt . grid ()
plt . show ()
```

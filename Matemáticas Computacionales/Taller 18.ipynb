```python
#Lineal
import numpy as np
from scipy import stats
import matplotlib . pyplot as plt

x = np . array ([0 ,1 ,2 ,3 ,4 ,5 ,6 ,7])
y = np . array ([3.5 ,2.5 ,3.0 ,1.5 ,2.0 ,1.3 ,1.0 ,0.3])

m , b , r , p , std_err = stats . linregress (x , y )

print ('Ordenada al origen ', b )
print ('pendiente ', m )
print ('coeficiente de correlacion ', r )

y_pred = m * x + b
print ('datos corregidos ', y_pred , sep ='\n')

plt . plot (x ,y ,'o', label ='Datos ')
plt . plot (x , y_pred ,'r:', label ='Regresion lineal ')
plt . title ('Taller 16')
plt . xlabel ('X ')
plt . ylabel ('Y ')
plt . legend ()
plt . grid ( True )
plt . show ()
```

```python
#Exponencial
import numpy as np
from scipy . optimize import curve_fit
import matplotlib . pyplot as plt

x = np . array ([2 ,4 ,6 ,8 ,10 ,12])
y = np . array ([2.2 ,3.0 ,4.5 ,6.0 ,8.5 ,12])

xs = np . linspace (0 ,15 ,15)

def f (x ,a , b ) :
  return a * np . exp ( - b * x )

param , _ = curve_fit (f ,x ,y , p0 =[0.1 , -0.0001])
print ( param )

fig = plt . figure ()
plt . scatter (x , y , label ='Datos ')
plt . plot ( xs , f ( xs ,* param ) , ':r', label ='Regresión exponencial ')
plt . xlabel ('X ')
plt . ylabel (' Y ')
plt . legend ()
plt . title ('Taller 17 ')
plt . grid ()
plt . show ()
```

```python
#Pontencias
import numpy as np
from scipy . optimize import curve_fit
import matplotlib . pyplot as plt
x = np . array ([0.5 ,1.5 ,2.5 ,3.5 ,4.5 ,5.5 ,6.5 ,7.5])
y = np . array ([4.2, 6.5, 7.5, 8.0, 8.5, 8.8 ,9.0 ,9.1])
xs = np . linspace (0 ,25 ,25)
def f (x ,a , b ) :
  return a * np . power (x , b )
param , _ = curve_fit (f ,x , y )
print ( param )

fig = plt . figure ()
plt . scatter (x , y , label ='Datos ')
plt . plot ( xs , f ( xs ,* param ) , ':r', label ='Regresión potencial ')
plt . xlabel ('Y ')
plt . ylabel ('X ')
plt . legend ()
plt . title ('Taller 18 ')
plt . grid ()
plt . show ()
```
